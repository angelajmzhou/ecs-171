Part 1) Use the following table and revise the probability values in the code below and include a conditional probability table based on the variables in this problem. Then Based on the structure of the model, provide the visualized Bayesian Network corresponding to the edges (i.e., dependencies) and the features (i.e., evidence).
In your computations, assume that P(grades=A) = 0.2, P(grades=B) = 0.4 , P(grades=C)=0.4, P(diff=easy)=0.5, P(diff=hard)=0.5, P(aptitude=low)=0.3, P(aptitude=medium) = 0.4, P(aptitude=high) = 0.3. 

![Generated Dataset](lab4-cpt.jpg)

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete.CPD import TabularCPD
import networkx as nx
import matplotlib.pyplot as plt


student = BayesianNetwork([('diff', 'grades'), ('aptitude', 'grades')])

cpd_aptitude = TabularCPD(
    variable='aptitude', 
    variable_card=3, 
    values=[[0.3], [0.4], [0.3]], 
    state_names={'aptitude': ['low', 'medium', 'high']}
)

# Define CPT for Difficulty (assuming uniform distribution)
cpd_difficulty = TabularCPD(
    variable='diff', 
    variable_card=2, 
    values=[[0.5], [0.5]], 
    state_names={'diff': ['easy', 'hard']}
)

grades_cpd = TabularCPD('grades', 3, [[0.1,0.2,0.5,0.05,0.1,0.3],
                                      [0.2,0.3,0.3,0.15,0.2,0.4],
                                      [0.7,0.5,0.2,0.8,0.7,0.3]],
                        evidence=['diff', 'aptitude'], evidence_card=[2, 3],
                        state_names={'grades': ['gradeA', 'gradeB', 'gradeC'],
                                     'diff': ['easy', 'hard'],
                                     'aptitude': ['low', 'medium', 'high']})
student.add_cpds(cpd_aptitude, cpd_difficulty, grades_cpd)



# Manually create a networkx graph from the BayesianNetwork
graph = nx.DiGraph()  # Directed graph for Bayesian network
graph.add_edges_from(student.edges())

# Visualize the network using circular layout
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(graph)  # Using circular layout as an alternative
nx.draw(graph, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=12, font_weight='bold', edge_color='red')
plt.title("Bayesian Network for Student Grades")
plt.show()
```
