What is the difference between Gradient Descent and Ordinary Lease Squares method? Explain the advantages and disadvantages of each approach.
***

Ordinary Least Squares methods is a mathematical method which obtains the global minima of the loss function. It's computationally expensive and doesn't work for larger datasets. However, this only works if the matrices' features are not colinear, as we need to be able to invert the matrices. Weights are found via the below formula: 
$$
w = X^T Y (X^T X)^{-1} 
$$
Meanwhile, the gradient descent method is more flexible and scalable, allowing for iterative calculation of a local minima in the lost function. A weight update rule is used to apply weight changes, and may take many cycles/epochs before converting to that local minimum. It's more efficient, but less accurate than OLS.


#### Question 4.2 - (2 points)
How does a negative or positive gradient indicate the direction towards achieving a sub-optimal or optimal weight? Explain your answer.
***

In the simplest example, the loss function of a Linear Regression model is quadratic, creating a "bowl-shaped" curve. We know this curve must have a minimum point because, no matter where we start with random weights, there's always an optimal solution somewhere along the curve. By calculating the gradient at our current point on the curve, we can determine the slope of the loss function at that point.

The gradient tells us whether we're moving towards or away from the minimum. A positive gradient indicates that we're heading in a direction where the error increases, meaning were moving away from the minimum. To correct this, weights should be adjusted in the opposite direction of the gradient (i.e. decrease the weight), as it leads us toward lower error. On the other hand, a negative gradient means that were already moving in a direction where the error is decreasing. In this case, we continue moving in the same direction as the gradient (increase the weight), guiding us closer to the local minimum.

#### Question 4.3 - (10 points)
Provide the complete math for updating Wj associated with the jth attribute of a dataset using Gradient Descent weight update rule for finding the coefficients of a linear regression model. Use the Sum of Squared Residuals (SSR) as the loss (or cost) function to compute the gradient (i.e., weight change). Provide all the necessary details explaining the parameters used in your work.
***

## Defining Variables
- $ n $ data points.
- $ X = [x_1, x_2, \dots, x_n] $ : input feature vectors, each having $ m $ attributes (features).
- $ y = [y_1, y_2, \dots, y_n] $ : vector of target values.
-  $ W = [W_0, W_1, \dots, W_m] $ are the model weights (coefficients).
- $\epsilon$ is the error, the difference between the model output and the actual output.

$ \hat{y}_j $ is predicted via linear regression for each data point according to the formula below:
$$
\hat{y}_j = W_0 x_{0} + W_1 x_{1} + W_2 x_{2} + \dots + W_m x_{m} + \epsilon
$$
- for simplicity, $x_{0}$ is set to 1. 

The **SSR** loss function for linear regression is:
$$
Loss = L(W) = \sum_{j=1}^{n} \left( y_i - \hat{y}_i \right)^2
$$
Where \( y_i \) is the actual value and \( \hat{y}_i \) is the predicted value for the \( i \)-th data point. Note that we square the values to prevent positive and negative residuals from cancelling each other out. 

Since differentiation finds the point at which a function is (at least locally) minimum, we must calculate the partial derivative of the loss function $ L(W) $ with respect to \( W_j \) to update $ W_j $. 

$$
\frac{\partial RSS}{\partial W_j} = \frac{\partial}{\partial W_j} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
$$
For one datapoint, and given that $\hat{y}_i = w^T x^{(i)}$ this is:
$$
RSS =\left( y_i - w^T x^{(i)} \right)^2
$$
And after taking the derivative and applying the power and chain rules:
$$
\frac{\partial RSS}{\partial W_j} = -2 * x_j^{(i)} * (y^{(i)} - wx^{(i)})
$$

For the reasons mentioned above, we want to decrease w if the gradient is greater than 0, and incresae w if the gradient is less than 0 to minimize the RSS.
We can do so with the formula below:

$$
W_j^{\text{new}} = W_j^{\text{old}} - \alpha \frac{\partial RSS}{\partial W_j}
$$

Where:
- $ \alpha $ is the learning rate (a small positive value controlling the step size).
- $\frac{\partial RSS}{\partial W_j}$ is the gradient computed earlier
- $W_j^{\text{old}}$ is the current model parameter corresponding to the jth dataset input attribute
- $W_j^{\text{new}}$ is the updated model parameter corresponding to the jth dataset input attribute

Substituting the gradient expression:
$$
W_j^{\text{new}} = W_j^{\text{old}} + 2 \alpha \frac {\partial \left( y_i - w^T x^{(i)} \right)^2}{\partial w_j}
$$

$$
W_j^{\text{new}} = W_j^{\text{old}} + 2 \alpha \frac { \partial \left( y^{(i)} - \sum_{k=0}^{m} w_k x_k^{(i)}\right)^2}{\partial w_j}
$$

(with $n$ being the total number of input attributes used to train the model)

$$
W_j^{\text{new}} = W_j^{\text{old}} + 2  \alpha \left( y^{(i)} - \sum_{k=0}^{m} w_k x_k^{(i)}\right) x_j^{(i)}
$$

What gradient descent does is iteratively updates each $ W_j $ using the above update rule until convergence -- when the weights stop changing, or those changes become very small (which indicates a minimum has been reached. The learning rate, or $\alpha$ plays a role in how fast the model converges, if at all. 


#### Question 4.4 - (5 points)
Explain the strategy in and the difference between stochastic gradient descent and batch-gradient descent algorithmically. 
***

Stochastic gradient descent iteratively updates the parameters of the model for every single randomly selected data point using the Gradient Descent weight update rule noted above. After all data points are updated, an epoch is complete, and the model's overall loss is measured. This is repeated until the model ultimately converges.  This can take many epochs to converge, or not converge at all, due to each individual data point playing "tug of war" with the weights. Here, the number of batches is equal to the number of observations.

Meanwhile, batch-gradient descent doesn't update weights for every data point; rather, it aggregates a batch of data points, updating the weights of the model with the average of the computed gradients after each batch. The number of the samples depends on the number of batches, each containing an equal partition of the dataset (depending on the batch size). This converges fast.

```python
from sklearn.linear_model import SGDRegressor

data = fetch_california_housing() #get the data
x = data.data #get the features
y = data.target #get the target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #20:80 test:train split, seeded with 42 for consistency

scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train) #normalize the training dataset with a z-score transformation
x_test_norm = scaler.transform(x_test) #normalize the testing dataset with a z-score transformation

#intialize a SGD regressor model: default squared error loss function, default l2 penalty/regularization term (standard for linear SVM)
#alpha - 0.0001, fit_intercept = true (default, estimates data intercept), tol: stopping criterion, set to none so it only stops after 100 epochs
#shuffle - true by default, verbose is false, random_state = 42 for consistent shuffling, learning rate is constant (stays 0.0001)

sgd_reg = SGDRegressor(tol=None, warm_start=True, learning_rate="constant", eta0 = 0.0001, random_state=42)

#intialize lists to hold data we want to plot
mse_train = []
mse_test = []
r2_train = []
r2_test = []

for epoch in range(100):
    #do one epoch of stochastic gradient descent. in the first pass, since no sample weights are provided, uniform weights are assumed.
    sgd_reg.partial_fit(x_train_norm, y_train)

    #based on those weights, predict y_train and y_test based on x_train and x_test
    y_train_pred = sgd_reg.predict(x_train_norm)
    y_test_pred = sgd_reg.predict(x_test_norm)

    #use library functions to calculate MSE between the ground truth and the predictions to graph later
    mse_train.append(mean_squared_error(y_train, y_train_pred))
    mse_test.append(mean_squared_error(y_test, y_test_pred))
    r2_train.append(r2_score(y_train, y_train_pred))
    r2_test.append(r2_score(y_test, y_test_pred))

#set up plot
plt.figure(figsize=(15,6))

# Plot MSE
plt.subplot(1, 2, 1)

#plot MSE for training and testing for each epoch
plt.plot(range(100), mse_train, label="MSE - Train")
plt.plot(range(100), mse_test, label="MSE - Test")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.title("MSE over 100 Epochs")
plt.legend()

# Plot R2
#plot R2 for training and testing for each epoch
plt.subplot(1, 2, 2)
plt.plot(range(100), r2_train, label="R2 - Train")
plt.plot(range(100), r2_test, label="R2 - Test")
plt.xlabel("Epoch")
plt.ylabel("R2 Score")
plt.title("R2 Score over 100 Epochs")
plt.legend()

plt.tight_layout()
plt.show()

```
